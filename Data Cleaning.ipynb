```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import os

import string
import nltk
from nltk.corpus import stopwords
```

```python
cols = ['Number','Title','Statement','Difficulty']
data = pd.DataFrame(columns = cols)

cur_tags = []

for path in os.listdir("problems"):
    file = open("problems/"+path, encoding="utf8") 
    soup = BeautifulSoup(file)
    
    results = soup.find(id = "sidebar").find_all("span", class_= "tag-box")

    tags = []

    for item in results:
        tags.append(item.string.split('\n')[1].strip())

    title = soup.find("div", class_="problem-statement").find_all("div", class_="title")[0].text.strip()
    
    statement_arr = soup.find("div", class_="problem-statement").find_all("div", class_=None)
    statement = ""
    for i in range(len(statement_arr)):
        statement+=' '.join(statement_arr[i].text.split())+' '
    statement = statement.strip()
    
    arr = [path[:-5], title, statement, tags[-1][1:]]
    for i in range(len(cur_tags)):
        arr.append(False)
    
    d = pd.DataFrame(arr).T
    d.columns = cols+cur_tags
    
    for tag in tags[:-1]:
        if tag in cur_tags:
            d[tag] = True
        else:
            cur_tags.append(tag)
            data[tag] = False
            d[tag] = True
            
    data = pd.concat([data, d])

    
data.reset_index(inplace = True)
data
```

```python
data.drop(columns = ['index'], inplace = True)
data.iloc[:,1] = data.iloc[:,1].str[3:]

data
```

```python
data.to_excel('raw data.xlsx')
```

```python
words = stopwords.words('english')

for i in range(1,3):
    data.iloc[:,i] = data.iloc[:,i].str.lower().str.translate(str.maketrans(string.punctuation, ' ' * len(string.punctuation)))

for i in range(1,3):
    for j in range(data.shape[0]):
        arr = data.iloc[j,i].split()
        new_arr = []
        for k in range(len(arr)):
            if arr[k] not in words and len(arr[k])>1:
                new_arr.append(arr[k])
        
        text = ' '.join(new_arr)
        data.iloc[j,i] = ''.join([c for c in text if ord(c)<128])
        

                
data
```

```python
data.to_excel('cleaned data.xlsx')
```
