```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import os

import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

import enchant
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
cols = ['Number','Title','Statement','Difficulty']
data = pd.DataFrame(columns = cols)

cur_tags = []

for path in os.listdir("problems"):
    file = open("problems/"+path, encoding="utf8") 
    soup = BeautifulSoup(file)
    
    results = soup.find(id = "sidebar").find_all("span", class_= "tag-box")

    tags = []

    for item in results:
        tags.append(item.string.split('\n')[1].strip())

    title = soup.find("div", class_="problem-statement").find_all("div", class_="title")[0].text.strip()
    
    statement_arr = soup.find("div", class_="problem-statement").find_all("div", class_=None)
    statement = ""
    for i in range(len(statement_arr)):
        statement+=' '.join(statement_arr[i].text.split())+' '
    statement = statement.strip()
    
    arr = [path[:-5], title, statement, tags[-1][1:]]
    for i in range(len(cur_tags)):
        arr.append(False)
    
    d = pd.DataFrame(arr).T
    d.columns = cols+cur_tags
    
    for tag in tags[:-1]:
        if tag in cur_tags:
            d[tag] = True
        else:
            cur_tags.append(tag)
            data[tag] = False
            d[tag] = True
            
    data = pd.concat([data, d])

    
data.reset_index(inplace = True)
data
```

```python
data.drop(columns = ['index'], inplace = True)
data.iloc[:,1] = data.iloc[:,1].str[3:]

data
```

```python
data.to_excel('raw data.xlsx')
```

```python
data.insert(3, "Raw Statement", data['Statement'])
data
```

```python
def pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:         
        return None
```

```python
d1 = enchant.Dict("en_US")
d2 = enchant.Dict("en_GB")

words = stopwords.words('english')

lemmatizer = WordNetLemmatizer()

data.iloc[:,2] = data.iloc[:,2].str.lower().str.translate(str.maketrans(string.punctuation, ' ' * len(string.punctuation)))
data.iloc[:,2] = data.iloc[:,2].str.translate(str.maketrans(string.digits, ' ' * len(string.digits)))

for j in range(data.shape[0]):
    arr = data.iloc[j,2].split()
    new_arr = []
    for k in range(len(arr)):
        if arr[k] not in words and len(arr[k])>1 and(d1.check(arr[k]) or d2.check(arr[k])):
            p = pos_tagger(nltk.pos_tag([arr[k]])[0][1])
            if(p==None):
                 new_arr.append(lemmatizer.lemmatize(arr[k]))
            else:
                new_arr.append(lemmatizer.lemmatize(arr[k], p))
        
    text = ' '.join(new_arr)
    data.iloc[j,2] = ''.join([c for c in text if ord(c)<128])
                
data
```

```python
d1 = enchant.Dict("en_US")
d2 = enchant.Dict("en_GB")


data.iloc[:,3] = data.iloc[:,3].str.lower().str.translate(str.maketrans(string.punctuation, ' ' * len(string.punctuation)))
data.iloc[:,3] = data.iloc[:,3].str.translate(str.maketrans(string.digits, ' ' * len(string.digits)))

for j in range(data.shape[0]):
    arr = data.iloc[j,3].split()
    new_arr = []
    for k in range(len(arr)):
        if len(arr[k])>1 and(d1.check(arr[k]) or d2.check(arr[k])):
            new_arr.append(arr[k])
            
        
    text = ' '.join(new_arr)
    data.iloc[j,3] = ''.join([c for c in text if ord(c)<128])
                
data
```
